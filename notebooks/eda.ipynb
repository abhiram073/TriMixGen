# Phase 3: Exploratory Data Analysis (EDA)

This notebook performs EDA on the TriMixGen datasets: `IndicCMix`, `DravidianCodeMix`, and `Bhasha-Abhijnaanam`.

```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import JSON
```

## 1. Load Precomputed EDA Report
We have a utility script `run_eda.py` that computes statistics across all datasets and saves them into `outputs/eda/full_eda_report.json`.

```python
report_path = '../outputs/eda/full_eda_report.json'
with open(report_path, 'r') as f:
    eda_report = json.load(f)
    
JSON(eda_report)
```

## 2. Visualizations
Let's display the publication-quality plots generated by our `DatasetAnalyzer`.

```python
from IPython.display import Image, display

datasets = ['IndicCMix', 'DravidianCodeMix', 'Bhasha-Abhijnaanam']
for ds in datasets:
    print(f"--- {ds} Visualizations ---")
    len_path = f'../outputs/eda/{ds}/sentence_length_dist.png'
    top_path = f'../outputs/eda/{ds}/top_words.png'
    
    if os.path.exists(len_path):
        display(Image(filename=len_path))
    if os.path.exists(top_path):
        display(Image(filename=top_path))
    print("\n")
```
